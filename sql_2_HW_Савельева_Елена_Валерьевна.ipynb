### Все задания нужно выполнить только с помощью sql

1. Выведите количество пустых значений по колонкам CRIM, ZN, INDUS, CHAS, NOX (название колонки, кол-во пустых значений)

```python
pd.read_sql(

'''

SELECT
(SELECT count(CRIM) from boston where CRIM = 0.0 OR CRIM is NULL) as 'crim',
(SELECT count(ZN) from boston where ZN = 0.0 OR ZN is NULL) as 'zn',
(SELECT count(INDUS) from boston where INDUS = 0.0 OR INDUS is NULL) as 'indus',
(SELECT count(CHAS) from boston where CHAS = 0.0 OR CHAS is NULL) as 'chas',
(SELECT count(NOX) from boston where NOX = 0.0 OR NOX is NULL) as 'nox'
FROM boston
GROUP BY
(SELECT count(CRIM) from boston where CRIM = 0.0 OR CRIM is NULL),
(SELECT count(ZN) from boston where ZN = 0.0 OR ZN is NULL),
(SELECT count(INDUS) from boston where INDUS = 0.0 OR INDUS is NULL),
(SELECT count(CHAS) from boston where CHAS = 0.0 OR CHAS is NULL),
(SELECT count(NOX) from boston where NOX = 0.0 OR NOX is NULL)

''',

con,

)
```

2. Выведите количество уникальных значений по колонокам CRIM, ZN, INDUS, CHAS, NOX (название колонки, кол-во уникальных значений)

```python
pd.read_sql(

'''

SELECT
(SELECT count(distinct CRIM) from boston) as 'dst_crim',
(SELECT count(distinct ZN) from boston) as 'dst_zn',
(SELECT count(distinct INDUS) from boston) as 'dst_indus',
(SELECT count(distinct CHAS) from boston) as 'dst_chas',
(SELECT count(distinct NOX) from boston) as 'dst_nox'
FROM boston
GROUP BY
(SELECT CRIM from boston),
(SELECT ZN from boston),
(SELECT INDUS from boston),
(SELECT CHAS from boston),
(SELECT NOX from boston)

''',

con,

)
```

4. Выведите ранги значений колонки LSTAT(процент населения с более низким статусом) в домах с самой дорогой стоимостью (значение LSTAT, стоимость, ранг) среди всех значений LSTAT. Напишите какой вывод можно сделать по этим данным.

```python
pd.read_sql(

'''

SELECT
LSTAT,MEDV,
RANK() OVER (ORDER BY MEDV DESC) as 'rnk_lstat'
FROM boston

--Вывод: более дорогую недвижимость покупает меньший процент населения.

''',
con,
)
```

5. Выведите среднюю стоимость домов граничащих с рекой(CHAS) и нет (граничит/не граничит, стоимость)

6. Выведите все колонки, у которых среднее значение выше, когда дом граничит с рекой (название колонки) выбирая из CRIM, ZN, INDUS, CHAS, NOX. Напишите какой вывод можно сделать по этим данным.

```python
pd.read_sql(

'''

SELECT
AVG(CRIM) OVER() as 'avg_crim',
AVG(ZN) OVER() as 'avg_zn',
AVG(INDUS) OVER() as 'avg_indus',
CASE
WHEN CHAS=1 THEN 'Граничит' ELSE 'Не граничит'
END AS ch
FROM boston
where ch='Граничит'
GROUP BY 'avg_crim','avg_zn','avg_indus','ch'
UNION ALL
SELECT
AVG(CRIM) OVER() as 'avg_crim',
AVG(ZN) OVER() as 'avg_zn',
AVG(INDUS) OVER() as 'avg_indus',
CASE
WHEN CHAS=1 THEN 'Граничит' ELSE 'Не граничит'
END AS ch
FROM boston
where ch='Не граничит'
GROUP BY 'avg_crim','avg_zn','avg_indus','ch'


-- Вывод:
-- Уровень преступности выше в домах, граничащих с рекой
-- Дома, граничащие с рекой меньше площади 25 000 кв. футов
-- Промышленные предприятия в большинстве случаев расположены у реки

''',
con,
)
```

7. Выведите значения долей промышленной застройки(INDUS), концентрации оксидов азота(NOX) и по их перцентилям - 10, 20 ... 100 ( перцетиль(10,20...100),значение INDUS, значение NOX). Напишите прослеживается между ними взаимосвязь

```python

pd.read_sql(

'''

SELECT count(INDUS)
FROM boston
where(
SELECT INDUS,
SUM(INDUS) OVER() as 'sum_indus',
sum_indus*100.0/indus OVER() as 'pr_indus',
FROM boston
where pr_indus < 10
)
UNION aLL
SELECT
count (NOX) fom boston
where (
SELECT NOX,
SUM(NOX) OVER() as 'sum_nox',
sum_nox*100.0/NOX OVER() as 'pr_nox',
FROM boston
where pr_indus < 10
)

''',
con,
)
```
